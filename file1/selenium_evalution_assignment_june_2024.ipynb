In this question you have to scrape data using the filters available on the webpage You have to use the location and
salary filter.
You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results.
You have to scrape the job-title, job-location, company name, experience required.
The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs
The task will be done as shown in the below steps:
1. first get the web page https://www.naukri.com/
2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, and Companies‚Äù field.
3. Then click the search button.
4. Then apply the location filter and salary filter by checking the respective boxes
5. Then scrape the data for the first 10 jobs results you get.
6. Finally create a dataframe of the scraped data.

```python
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
jobtitle = []
joblocation = []
companyname = []
experiencerequired = []
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/")#to get webpage
```

```python
#enter data scientist using absolute xpath
position = driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input")
driver.implicitly_wait(2)
position.send_keys("Data Scientist")
```

```python
#clicked on search bar
search = driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[6]")
search.click()
```

```python
location = driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[3]/label/i")
location.click()
```

```python
salary = driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[6]/div[2]/div[2]/label/i")
salary.click()
```

```python
job_tags = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in job_tags:
    t = i.text
    jobtitle.append(t)
```

```python
jobtitle
```

```python
location_tag = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc"]')
for i in location_tag:
    joblocation.append(i.text)
```

```python
joblocation
```

```python
company_tags = driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
driver.implicitly_wait(2)
for i in company_tags:
    companyname.append(i.text)
```

```python
companyname
```

```python
exeperience_tags = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp"]')
driver.implicitly_wait(2)
for i in exeperience_tags:
    experiencerequired.append(i.text)
experiencerequired
```

```python
df = pd.DataFrame({'job-title':jobtitle[:10], 'job-location':joblocation[:10], 'company name':companyname[:10], 'experience required':experiencerequired[:10]})
```

```python
quote = []
for page in range(0,10):
    quote_tag = driver6.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tag:
        quote.append(i.text)
    next_button = driver6.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(5)
```

```python
len(quote)
quote
```

```python
Author = []
for page in range(0,10):
    Author_tag = driver6.find_elements(By.XPATH,'//div[@class="author"]')
    for i in Author_tag:
        Author.append(i.text)
    next_button = driver6.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(5)
```

```python
len(Author)
```

```python
Type = []
for page in range(0,10):
    Type_tag = driver6.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in Type_tag:
        Type.append(i.text)
    next_button = driver6.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(4)
```

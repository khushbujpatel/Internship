have to find the following details:
    
A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute

G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver8 = webdriver.Chrome()
```

```python
driver8.get("https://archive.ics.uci.edu/")
```

```python
viewdataset = driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
viewdataset.click()
```

```python
set_url =[]
while len(set_url)< 665:
    for i in driver8.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]'):
        set_url.append(i.get_attribute("href"))
    nxt_button = driver8.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    nxt_button.click()
```

```python
len(set_url)
```

```python
Datatype =[]
Dataset_name = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = []
Year = []
for i in set_url[:20]:
    driver8.get(i)
    try:
        Datasetname_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1")
        Dataset_name.append(Datasetname_tag.text)
    except NoSuchElementException as e:
        Dataset_name.append("--")
    try:
        Datatype_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/h1")
        Datatype.append(Datatype_tag.text)
    except NoSuchElementException as e:
        Datatype.append("--")
    try:
        Task_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p")
        Task.append(Task_tag.text)
    except NoSuchElementException as e:
        Task.append("--")
    try:
        Attribute_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p")
        Attribute_type.append(Attribute_tag.text)
    except NoSuchElementException as e:
        Attribute_type.append("--")
    try:
        No_of_instances_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p")
        No_of_instances.append(No_of_instances_tag.text)
    except NoSuchElementException as e:
        No_of_instances.append("--")
    try:
        No_of_attribute_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p")
        No_of_attribute.append(No_of_attribute_tag.text)
    except NoSuchElementException as e:
        No_of_attribute.append("--")
    try:
        Year_tag=driver8.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2")
        Year.append(Year_tag.text)
    except NoSuchElementException as e:
        Year.append("--")
```

```python
len(Dataset_name)
```

```python
df8 = pd.DataFrame({"Datatype":Datatype,"Dataset name":Dataset_name,"Task":Task,"Attribute type":Attribute_type,"No_of_instances":No_of_instances,"No_of_attribute":No_of_attribute,"Year":Year})
```

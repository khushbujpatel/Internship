2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year
of release) and make data frame.
https://m.imdb.com/chart/top/

```python
#page_2 = requests.get("https://m.imdb.com/chart/top/")
#page_2
import requests

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'} # This is chrome, you can set whatever browser you like
page_2 = requests.get('https://m.imdb.com/chart/top/', headers=headers)

print(page_2.status_code)
print(page_2.url)
```

```python
soup_2 = BeautifulSoup(page_2.content)
```

```python
soup_2
```

```python
Movie_Name = []
for i in soup_2.find_all('a',class_="ipc-title-link-wrapper",limit=100):
    Movie_Name.append(i.text)
```

```python
Movie_Name
```

```python
rating = []
import regex as re
for i in soup_2.find_all('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating",limit=100):
    result = re.sub(r"\xa0","",i.text) 
    rating.append(result)
```

```python
rating
```

```python
import regex as re
year = []
release = []
for i in soup_2.find_all('div',class_="sc-b189961a-7 feoqjK cli-title-metadata",limit=100):
    release = re.findall('^\d{4}',i.text)
    year.append(release)
```

```python
year
```

```python
import pandas as pd
df = pd.DataFrame({"Name":Movie_Name,"Rating":rating,"Release year":year})
df
```

```python
page_3
```

```python
soup_3 = BeautifulSoup(page_3.content)
soup_3
```

```python
name = []
for i in soup_3.find_all('a',class_= "restnt-name ellipsis"):
    name.append(i.text)
```

```python
name
```

```python
Cuisine = []
for i in soup_3.find_all('div',class_= "detail-info"):
    c = re.findall(r"\|.+",i.text)
    Cuisine.append(c)
Cuisine
```

```python
loc = []
for i in soup_3.find_all("div",class_="restnt-loc ellipsis"):
    loc.append.(i.text)
```
